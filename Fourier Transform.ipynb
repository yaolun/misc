```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-15,15,30./100)
signal = np.sin(2*np.pi*0.7*x) + np.sin(2*np.pi*x)

fourier = np.fft.fft(signal)
freq = np.fft.fftfreq(len(signal), d=30/100.)
```

```python
plt.plot(x, signal)
```

```python
plt.plot(freq, fourier)
# plt.xlim([0,2])
```
