```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
home = os.path.expanduser('~')
# Constant Setup
#
AU        = 1.49598e13     # Astronomical Unit       [cm]
pc        = 3.08572e18     # Parsec                  [cm]
MS        = 1.98892e33     # Solar mass              [g]
LS        = 3.8525e33      # Solar luminosity        [erg/s]
RS        = 6.96e10        # Solar radius            [cm]
G         = 6.67259e-8     # Gravitational constant  [cm3/g/s^2]
yr        = 60*60*24*365   # Years in seconds
PI        = np.pi          # PI constant
```

```python
data = np.genfromtxt(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/vrot.txt',dtype='float')
r = data[:,0]
vc = data[:,1]
```

```python
# circular velocity
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(np.log10(r),vc,linewidth=1.5)
ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{V_{c}~[km/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_vc.eps',format='eps',dpi=300,bbox_inches='tight')
# fig.clf()
```

```python
# Angular velocity
omega = vc/r/pc*1e2
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(np.log10(r),omega,linewidth=1.5)
ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{\Omega~[rad/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_omega.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
# epicyclic frequency
kappa1 = (r[0:-1]*np.diff(omega**2)/np.diff(r)+4*(omega[0:-1])**2)**0.5
kappa2 = (2*(vc/r)[0:-1]*((vc/r)[0:-1]+np.diff(vc)/np.diff(r)))**0.5 /pc*1e2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
k1, = ax.plot(np.log10(r[0:-1]),kappa1,'b-',linewidth=1.5)
k2, = ax.plot(np.log10(r[0:-1]),kappa2,'r-',linewidth=1.5)
ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{\kappa~[rad/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
lg = plt.legend([k1,k2],[r'$\mathrm{\kappa^{2}=R\frac{d\Omega^{2}}{dR}+4\Omega^{2}}$',\
                         r'$\mathrm{\kappa^{2}=\frac{2V_{c}}{R}\left(\frac{V_{c}}{R}+\frac{dV_{c}}{dR}\right)}$'],\
                fontsize=18)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_kappa.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
# ultraharmonic resonance
uhr = omega[0:-1]-kappa1/4.

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(np.log10(r[0:-1]),uhr,linewidth=1.5)
ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{(\Omega-\kappa/4)~[rad/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_uhr.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
# inner Lindblad resonance
ilr = omega[0:-1]-kappa1/2.

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(np.log10(r[0:-1]),ilr,linewidth=1.5)
ax.axhline(37.9/pc*1e2,color='r')
ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{(\Omega-\kappa/2)~[rad/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_ilr.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
# All kinds of frequencies in one plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
omega_plot, = ax.plot(np.log10(r[0:-3]),omega[0:-3],linewidth=1.5)
k1, = ax.plot(np.log10(r[0:-1]),kappa1,linewidth=1.5)
k2, = ax.plot(np.log10(r[0:-1]),kappa2,linewidth=1.5)
uhr_plot, = ax.plot(np.log10(r[0:-1]),uhr,linewidth=1.5)
ilr_plot, = ax.plot(np.log10(r[0:-1]),ilr,linewidth=1.5)

ax.set_xlabel(r'$\mathrm{log(R)~[kpc]}$',fontsize=20)
ax.set_ylabel(r'$\mathrm{Frequency~[rad/s]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
lg = plt.legend([omega_plot,k1,k2,uhr_plot,ilr_plot],[r'$\mathrm{angular~frequency}$',\
                         r'$\mathrm{\kappa^{2}=R\frac{d\Omega^{2}}{dR}+4\Omega^{2}}$',\
                         r'$\mathrm{\kappa^{2}=\frac{2V_{c}}{R}\left(\frac{V_{c}}{R}+\frac{dV_{c}}{dR}\right)}$',\
                         r'$\mathrm{(\Omega-\kappa/4)}$',\
                         r'$\mathrm{(\Omega-\kappa/2)}$'],fontsize=18)
fig.savefig(home+'/Dropbox/UT-Austin/Courses/AST386c_Galaxies/hw3_all.eps',format='eps',dpi=300,bbox_inches='tight')
```
