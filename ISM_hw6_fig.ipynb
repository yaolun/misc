```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const
pc = const.pc.cgs.value
au = const.au.cgs.value
mag = 1.5

# r = np.arange(0.001,100,0.0001)  # radius in parsec
# r = r*pc/au
# rate = 14.1*(-10*r**-2+331.5*r**(-7./3))

r = np.arange(0.001,1000,0.0001)  # radius in 100 AU
T_gas = 10.0
rate = 9e-34*2.66**2*1e12*r**-4*T_gas**0.5*(1-0.8*np.exp(-75.0/T_gas))*(T_gas-56.7/r**(1/3.))
rate_cr = 1e-27*1e6*r**-2

fig = plt.figure(figsize=(8*mag,6*mag))
ax = fig.add_subplot(111)
dust_heat, = ax.plot(np.log10(r[rate < 0]),np.log10(-rate[rate < 0]),'b-',linewidth=1.5*mag)
dust_cool, = ax.plot(np.log10(r[rate > 0]),np.log10(rate[rate > 0]),'b--',linewidth=1.5*mag)
cr_heat, = ax.plot(np.log10(r),np.log10(rate_cr),'r-',linewidth=1.5*mag)
lg = plt.legend([dust_heat,dust_cool,cr_heat],\
                [r'$\mathrm{Dust~Heating}$',r'$\mathrm{Dust~Cooling}$',r'$\mathrm{Cosmic~Ray~Heating}$'],\
                loc = 'upper right',fontsize=17*mag)
ax.set_xlabel(r'$\mathrm{log(r)~[100~AU]}$',fontsize=19*mag)
ax.set_ylabel(r'$\mathrm{log(Collisional~Heating~Rate)~[ergs~cm^{-3}~s^{-1}]}$',fontsize=19*mag)
[ax.spines[axis].set_linewidth(1.5*mag) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=mag*18,width=1.5*mag,which='major',pad=15,length=5*mag)
ax.tick_params('both',labelsize=mag*18,width=1.5*mag,which='minor',pad=15,length=2.5*mag)
fig.savefig('/Users/yaolun/Dropbox/UT-Austin/Courses/AST393_ISM/hw6_q2_fig.eps',format='eps',dpi=300,bbox_inches='tight')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const
pc = const.pc.cgs.value
au = const.au.cgs.value
mag = 1.5

r = np.arange(0.001,1000,0.0001)  # radius in 100 AU
nd = 1e6/r**2
ndeff = 1e6/r**2
ndeff[ndeff>1e8/np.pi/0.25]=1e8/np.pi/0.25
t_silicate = nd/nd*56.7/r**(1/3.)
t_bb = nd/nd*27.7/r**(1/2.)
fig = plt.figure(figsize=(8*mag,6*mag))
ax = fig.add_subplot(111)
silicate, = ax.plot(np.log10(r),np.log10(t_silicate),'b-',linewidth=1.5*mag)
bb, = ax.plot(np.log10(r),np.log10(t_bb),'r-',linewidth=1.5*mag)
lg = plt.legend([silicate,bb],['Astrosilicate','Black~Body~Dust'],loc='upper right',fontsize=17*mag)
ax.set_xlabel(r'$\mathrm{log(r)~[100~AU]}$',fontsize=19*mag)
ax.set_ylabel(r'$\mathrm{log(Temperature)~[K]}$',fontsize=19*mag)
[ax.spines[axis].set_linewidth(1.5*mag) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=mag*18,width=1.5*mag,which='major',pad=15,length=5*mag)
ax.tick_params('both',labelsize=mag*18,width=1.5*mag,which='minor',pad=15,length=2.5*mag)
fig.savefig('/Users/yaolun/Dropbox/UT-Austin/Courses/AST393_ISM/hw6_q2-1_fig.eps',format='eps',dpi=300,bbox_inches='tight')
```
