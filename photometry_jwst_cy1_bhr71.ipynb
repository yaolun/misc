```python
import matplotlib.pyplot as plt
from astropy.wcs import WCS
from astropy.io import fits
from mpl_toolkits.axes_grid1.axes_rgb import imshow_rgb
from astropy.visualization import MinMaxInterval, LogStretch, ManualInterval
from matplotlib.patches import Rectangle
import numpy as np

hdu = fits.open('/Users/yaolun/GoogleDrive/research/bhr71/data/J_IRAC1_250um.fits')
hdu[0].header['NAXIS'] = 2
hdu[0].header['NAXIS3'] = 1

wcs = WCS(hdu[0].header)

# slice the wcs and data
wcs = wcs[2420:4200, 2560:4340]
data = hdu[0].data[:,2420:4200, 2560:4340]

transform = LogStretch() + ManualInterval(vmax=1500)
r = transform(data[0])

transform = LogStretch() + ManualInterval(vmax=150)
g = transform(data[1])
transform = LogStretch() + ManualInterval(vmax=500)
b = transform(data[2])

fig = plt.figure()
ax = fig.add_subplot(111, projection=wcs)

imshow_rgb(ax, r, g, b, origin='lower')

# NIRCam FoV
nircam = Rectangle((center.ra.degree-2.2/60/np.cos(np.radians(center.dec.degree))/2,
                    center.dec.degree-2.2/60/2),
                   2.2/60/np.cos(np.radians(center.dec.degree)), 2.2/60, edgecolor='white', facecolor='none',
                   transform=ax.get_transform('fk5'))
ax.add_patch(nircam)

# set coordinates format
ax.coords[0].set_major_formatter('hh:mm:ss')
ax.coords[0].set_axislabel('Right Ascension (J2000)')

ax.coords[1].set_major_formatter('dd:mm:ss')
ax.coords[1].set_axislabel('Declination (J2000)')

fig.savefig('/Users/yaolun/GoogleDrive/proposal/JWST/BHR71_NIR/bhr71_j_irac1_spire250.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
