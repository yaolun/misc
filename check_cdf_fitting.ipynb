```python
def FittingSummary(filename):
    from astropy.io import ascii
    data = ascii.read(filename)
    
    # Get a list of object
    objlist = list(set(data['Object']))
    
    # Get a pixel counts for each object
    pixel_counts = []
    for o in objlist:
        
        pix_all = data['Pixel_No.'][data['Object'] == o]
        pix = list(set(pix_all))
        pixel_counts.append(len(pix))
        
        print o, list(set(data['RA(deg)'][data['Object'] == o]))
        
        # check if every pixel has the same amount of lines fitted
        line_counts = []
        for p in pix:
            line_counts.append(len(data['Line'][(data['Object'] == o) & (data['Pixel_No.'] == p)]))
        if len(set(line_counts)) > 2:
            print 'Line counts anomaly at ', o
            print pix
            print line_counts
    
    return objlist, pixel_counts
```
