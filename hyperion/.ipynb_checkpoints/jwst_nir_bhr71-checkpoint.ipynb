```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from hyperion.model import ModelOutput
import astropy.constants as const
from astropy.io import ascii
pc = const.pc.cgs.value
dstar = 200.0
```

```python
m = ModelOutput('/home/bettyjo/yaolun/hyperion/bhr71/model1/model1.rtout')
```

```python
wl_aper, aper_arcsec = np.genfromtxt('/home/bettyjo/yaolun/bhr71/best_calibrated/aperture.txt',
                                     skip_header=1, dtype=float).T
aper_sed = sorted(aper_arcsec)
```

```python
seds = [m.get_sed(group=i, inclination=0, aperture=-1, distance=dstar*pc, uncertainties=True) 
        for i in range(len(list(set(aper_arcsec)))+1)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# infinite aperture
ax.plot(np.log10(seds[0].wav), np.log10(seds[0].val), label='infinite aperture')

for sed, aperture in zip(seds[1:], aper_sed):
    ax.plot(np.log10(sed.wav), np.log10(sed.val), label='{:<.1f} arcsec'.format(aperture))

ax.legend(loc='best', fontsize=16)
    
ax.set_xlabel(r'log $\lambda$ [$\mu$m]', fontsize=18)
ax.set_ylabel(r'log $\nu S_{\nu}$ [erg s$^{-1}$ cm$^{-2}$]', fontsize=18)
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=6)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=4)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
```

### Further examine the 0.1 arcsec SED

```python
sed_100mas = m.get_sed(group=3, inclination=0, aperture=-1, distance=dstar*pc, uncertainties=True)

wav_select = (sed_100mas.wav >= 1.5) & (sed_100mas.wav <= 5)

plt.plot(sed_100mas.wav[wav_select], np.log10(sed_100mas.val[wav_select]), color='b')
plt.fill_between(sed_100mas.wav[wav_select],
                 np.log10(sed_100mas.val[wav_select]-sed_100mas.unc[wav_select]),
                 np.log10(sed_100mas.val[wav_select]+sed_100mas.unc[wav_select]),
                 color='b', alpha=0.3)
# plt.yscale('log')
```

```python
# plt.plot(sed_100mas.wav[sed_100mas.wav < 3], sed_100mas.val[sed_100mas.wav < 3])
plt.plot(sed_100mas.wav[wav_select], (sed_100mas.unc/sed_100mas.val)[wav_select])
```

### Analysis the 0.1 arcsec spectra from different models

```python
models = ['2','3','4']
ms = [ModelOutput('/home/bettyjo/yaolun/hyperion/bhr71/model'+model+'/model'+model+'.rtout') for model in models]
```

```python
wl_aper, aper_arcsec = np.genfromtxt('/home/bettyjo/yaolun/bhr71/best_calibrated/aperture.txt',
                                     skip_header=1, dtype=float).T
aper_sed = sorted(aper_arcsec)
aper_ind = aper_sed.index(0.1)
```

```python
seds = [m.get_sed(group=aper_ind+1, inclination=0, aperture=-1, distance=dstar*pc, uncertainties=True) 
        for m in ms]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

for sed, model in zip(seds[1:], models):
    ax.plot(np.log10(sed.wav), np.log10(sed.val), label='Model{:s}'.format(model))

ax.legend(loc='best', fontsize=16)
    
ax.set_xlabel(r'log $\lambda$ [$\mu$m]', fontsize=18)
ax.set_ylabel(r'log $\nu S_{\nu}$ [erg s$^{-1}$ cm$^{-2}$]', fontsize=18)
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=6)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=4)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
```
