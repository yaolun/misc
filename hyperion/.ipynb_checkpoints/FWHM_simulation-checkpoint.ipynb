```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import astropy.constants as const
from hyperion.model import ModelOutput

pc = const.pc.cgs.value
AU = const.au.cgs.value
dstar = 200.
wav = 1300.
```

```python
m = ModelOutput('/Volumes/SD-Mac/model24.rtout')
```

```python
wl_aper, aper_arcsec = np.genfromtxt('/Users/yaolun/bhr71/best_calibrated/aperture.txt', skip_header=1, dtype=float).T
aper_reduced = list(set(aper_arcsec))
```

```python
image = m.get_image(group=len(aper_reduced)+1, inclination=0, distance=dstar*pc, units='MJy/sr')
```

```python
iwav = np.argmin(np.abs(wav - image.wav))
val = image.val[::-1, :, iwav] + 1e-30

# get plate scale
rmax = max(m.get_quantities().r_wall)
w = np.degrees(rmax / image.distance) * 3600.
pix = len(val[:,0])
pix2arcsec = 2*w/pix
```

```python
# convolve with the beam size of the observation compared with
from astropy.convolution import convolve, Gaussian2DKernel
beam_size = 45  # arcsec, should be FWHM of the beam
kernel = Gaussian2DKernel(45/2.354/pix2arcsec)
val_convolved = convolve(val, kernel)
```

```python
# create index grid
grid_x, grid_y = np.meshgrid(np.linspace(0,pix-1,pix), np.linspace(0,pix-1,pix))
grid_x = grid_x - (pix-1)/2.
grid_y = grid_y - (pix-1)/2.
dist = (grid_x**2 + grid_y**2)**0.5

print(dist.max()*pix2arcsec*dstar*AU/pc)

# mask out the pixels with flux lower than half of the peak
dist_trim = np.where(val_convolved >= val_convolved.max()/2, dist, 0)
val_trim = np.where(val_convolved >= val_convolved.max()/2, val_convolved, 0)

print(dist_trim.max()*pix2arcsec*dstar*AU/pc)
```

```python
plt.imshow(np.log10(val_convolved), cmap='viridis')
```

```python
plt.imshow(np.log10(val_trim), cmap='viridis')
```
