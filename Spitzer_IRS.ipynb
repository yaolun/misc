```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
filename = '/Users/yaolun/data/Spitzer/IRAS03301+3111.tbl'
data = ascii.read(filename)
```

```python
data
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(data['wavelength'], data['flux_density'])
hcn = ax.axvline(12.776, color='k', linestyle='--')
water = ax.axvline(12.3765, color='k', linestyle=':')
ax.axvline(12.3971, color='k', linestyle=':')
ax.axvline(12.4079, color='k', linestyle=':')
ax.legend([hcn, water], [r'$\rm{HCN}$', r'$\rm{water}$'], fontsize=18, loc='best')
ax.set_xlim([12,13.2])
ax.set_xlabel(r'$\rm{Wavelength\,[\mu m]}$',fontsize=20)
ax.set_ylabel(r'$\rm{Flux\,Density\,[Jy]}$',fontsize=20)
[ax.spines[axis].set_linewidth(1.5) for axis in ['top','bottom','left','right']]
ax.minorticks_on()
ax.tick_params('both',labelsize=18,width=1.5,which='major',pad=15,length=5)
ax.tick_params('both',labelsize=18,width=1.5,which='minor',pad=15,length=2.5)
```
